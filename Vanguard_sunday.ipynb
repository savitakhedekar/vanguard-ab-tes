# < KPI ERRORS SUNDAY

```python
def check_eerrors(steps):
    steps_order = {
        "start": "step_1",
        "step_1": "step_2",
        "step_2": "step_3",
        "step_3": "confirm",
        "confirm": "",
    }
    
    # steps = digital_footprints_kpi_errors['process_steps']
    
    i_1 = 0
    i_2 = 1
    count_errors = 0
    for i in range(len(steps) - 1):
        if steps[i_2] != steps_order[steps[i_1]]:
            count_errors += 1
        i_1 += 1
        i_2 += 1
    
    if count_errors == 0 and len(steps) < 5:
        count_errors += 1

    return count_errors
```

```python
digital_footprints_kpi_errors['count_errors'] = digital_footprints_kpi_errors.apply(lambda row: check_eerrors(row['process_steps']), axis = 1)
```
